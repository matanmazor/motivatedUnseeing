```python
#-----------------------------------------------------------------------------#
# model 1: basemodel (nothing can vary by expCW and contingency)
def run_basemodel(id):
    import hddm
    import os
    data = hddm.load_csv('../experiments/bird_fish/data/forDDM.csv')
    # Daniel and Emily always set nrsamples to 30000 and didn't use thinning, so you could also try that
    nrsamples = 90000
    m1 = hddm.HDDMStimCoding(data,stim_col='stim',split_param='v',include='z',p_outlier=0.05,drift_criterion=True,group_only_nodes=[])
    m1.find_starting_values()
    m1.sample(nrsamples,burn=nrsamples/4,dbname='traces%i'%(id+1),db='pickle',thin=3)
    m1.save('DDM/basemodel/model1_%i'%(id+1))
    return True


run_basemodel(0)
```
